```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "Screenshot 2025-08-13 210836.png"
image = cv2.imread(image_path)
image = cv2.resize(image, (1920, 1080))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
dot_coordinates = [[110, 1035], [265, 400], [1600, 400], [1900, 915]]

# Loop through the coordinates and draw a filled circle (a dot) at each point
for coords in dot_coordinates:
    # The center is a tuple of (x, y)
    center = (coords[0], coords[1])
    # The radius of the dot in pixels
    radius = 10
    # Color of the dot in BGR format (e.g., green is (0, 255, 0))
    color = (0, 255, 0)
    # Use -1 for thickness to create a filled circle
    thickness = 5
    cv2.circle(image, center, radius, color, thickness)
```

```python
for i in range(len(dot_coordinates) - 1):
    # The first point of the line
    pt1 = tuple(dot_coordinates[i])
    # The second point of the line
    pt2 = tuple(dot_coordinates[i+1])
    # Color of the line in BGR format (e.g., red is (0, 0, 255))
    line_color = (0, 0, 255)
    # The thickness of the line in pixels
    line_thickness = 5
    cv2.line(image, pt1, pt2, line_color, line_thickness)


# Display the image with the dots in a window named "Image Display"
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
